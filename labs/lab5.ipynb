```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os, sys, re
import seaborn as sns

episodes = pd.read_csv('simpsons_episodes.csv')
```

```python
# extract some datetime variables
episodes['datetime'] = pd.to_datetime(episodes['original_air_date'])
episodes['datetime_year'] = episodes['datetime'].dt.year
episodes['datetime_month'] =episodes['datetime'].dt.month
episodes['datetime_day'] =episodes['datetime'].dt.day
episodes['datetime-MM-DD'] = episodes['datetime'].dt.strftime('%m-%d')
```

## Vizualizacija dviju kontinuiranih varijabli

- Je li se smanjivao IMDB rating i US views kroz godine
- Postoji mnogo načina kako se može ispitati odnos između dvije varijable. Najjednostavniji je dijagram scatterplot, koristeći plt.scatterplot()

```python
plt.figure(figsize=(20, 7))
sns.scatterplot(data=episodes, x="datetime", y="imdb_rating", alpha=0.6)
plt.xlabel("Years")
plt.ylabel("IMDB rating")
```

## Vizualizacija dviju varijabli, kontinuiranih i kategoričkih

- Razmotrimo IMDB rating i US views kroz godine i sezone emitiranja
- Prikazat ćemo barplot, boxplot i violinplot. Seaborn nudi mnogo više uvjerljivih vizualizacija. Bacite oko na primjere: https://seaborn.pydata.org/examples/index.html

```python
# barplot
fig, ax1 = plt.subplots(figsize=(20,7))

sns.barplot(data=episodes, x="datetime_year", y="imdb_rating", color="skyblue")
ax1.set_xlabel("Year")
ax1.set_ylabel("IMDB")
ax1.tick_params(axis="y", labelcolor="skyblue")

ax2 = ax1.twinx()
sns.barplot(data=episodes, x="datetime_year", y="views", color="orange")
ax2.set_xlabel("Year")
ax2.set_ylabel("US views")
ax2.tick_params(axis="y", labelcolor="orange")

fig, ax3 = plt.subplots(figsize=(20,7))

sns.barplot(data=episodes, x="season", y="imdb_rating", color="lightgreen")
ax3.set_xlabel("Seasons")
ax3.set_ylabel("IMDB")
ax3.tick_params(axis="y", labelcolor="lightgreen")

ax4 = ax3.twinx()
sns.barplot(data=episodes, x="season", y="views", color="purple")
ax4.set_xlabel("Seasons")
ax4.set_ylabel("US views")
ax4.tick_params(axis="y", labelcolor="purple")
```

```python
# boxplot
fig, ax1 = plt.subplots(figsize=(20,7))

sns.boxplot(data=episodes, x="datetime_year", y="imdb_rating", color="skyblue")
ax1.set_xlabel("Year")
ax1.set_ylabel("IMDB")
ax1.tick_params(axis="y", labelcolor="skyblue")

ax2 = ax1.twinx()
sns.boxplot(data=episodes, x="datetime_year", y="views", color="orange")
ax2.set_xlabel("Year")
ax2.set_ylabel("US views")
ax2.tick_params(axis="y", labelcolor="orange")

fig, ax3 = plt.subplots(figsize=(20,7))

sns.boxplot(data=episodes, x="season", y="imdb_rating", color="lightgreen")
ax3.set_xlabel("Seasons")
ax3.set_ylabel("IMDB")
ax3.tick_params(axis="y", labelcolor="lightgreen")

ax4 = ax3.twinx()
sns.boxplot(data=episodes, x="season", y="views", color="purple")
ax4.set_xlabel("Seasons")
ax4.set_ylabel("US views")
ax4.tick_params(axis="y", labelcolor="purple")
```

```python
# violinplot
fig, ax1 = plt.subplots(figsize=(20,7))

sns.violinplot(data=episodes, x="datetime_year", y="imdb_rating", color="skyblue")
ax1.set_xlabel("Year")
ax1.set_ylabel("IMDB")
ax1.tick_params(axis="y", labelcolor="skyblue")

ax2 = ax1.twinx()
sns.violinplot(data=episodes, x="datetime_year", y="views", color="orange")
ax2.set_xlabel("Year")
ax2.set_ylabel("US views")
ax2.tick_params(axis="y", labelcolor="orange")

fig, ax3 = plt.subplots(figsize=(20,7))

sns.violinplot(data=episodes, x="season", y="imdb_rating", color="lightgreen")
ax3.set_xlabel("Seasons")
ax3.set_ylabel("IMDB")
ax3.tick_params(axis="y", labelcolor="lightgreen")

ax4 = ax3.twinx()
sns.violinplot(data=episodes, x="season", y="views", color="purple")
ax4.set_xlabel("Seasons")
ax4.set_ylabel("US views")
ax4.tick_params(axis="y", labelcolor="purple")
```

## 2.1 Heatmape

- Vizualizirajte IMDB i US views kroz sezone te koristite funkciju Seaborn heatmap te funkciju *pivot* of pandasa

- Konačno, heatmapom možemo predstaviti tri varijable, kodiranjem treće varijable bojom. Pokažite kako sa sezonom opada IMDB ocjena i US views korištenjem heatmape


```python
# Napišite svoj kod ovdje da biste kreirali prvu heatmapu
data_matrix = episodes.pivot_table(index='number_in_season', columns='season', values='imdb_rating')
data_matrix

plt.figure(figsize=(14, 10))
sns.heatmap(data_matrix, annot=True, cmap='Blues', fmt='.1f', linewidth='.5', linecolor='white', cbar_kws={'label': 'IMDB Rating'})

plt.title('IMDB Rating per Episode', fontsize=16)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Episode Number in Season', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Napišite svoj kod ovdje da biste kreirali drugu heatmapu
data_matrix = episodes.pivot_table(index='number_in_season', columns='season', values='us_viewers_in_millions')
data_matrix

plt.figure(figsize=(14, 10))
sns.heatmap(data_matrix, annot=True, cmap='Blues', fmt='.1f', linewidth='0.7', linecolor='gray', cbar_kws={'label': 'US viewers in millions'})

plt.title('US viewers per episodes', fontsize=16)
plt.xlabel('Season', fontsize=12)
plt.ylabel('US viewers', fontsize=12)

plt.tight_layout()
plt.show()
```
